```python
import pandas as pd
```

```python
data = pd.read_csv("Hargovind - Sheet1.csv")
```

```python
data
```

```python
index_to_drop = data.index[data["Bill Date"].str.contains("Total")].to_list()
```

```python
for i in index_to_drop:
    data = data.drop(i)
data.reset_index(drop=True)
```

```python
data.drop(columns=["Old Bill Ref.", "Item Promo Code", "Bill Promo Code"])
```

```python
data["Profit"] = (data["MRP"]-data["Cost"])*data["Qty"]
```

```python
data.to_csv("Cleaned_Data.csv")
```

```python
profits = data.groupby(["Bill Date"]).sum()["Profit"]
```

```python
profits.to_csv("Profit.csv")
```

```python
profits.max()
```

```python
profits.min()
```

```python
profits.mean()
```
